```python
# cell 2
# reading the csv files
df1=pd.read_csv("AAPL.csv")
```

```python
# cell 5
# data preprocessing is carried out
# to carry out the analysis we need to drop the date column as it can't be converted to numerical value for analysis
df1_processed = df1.drop(columns=["Date","Date_col"])
```

```python
# cell 6
# creating the train and test set for the data
def create_train_test_dataset(df_s):
    features = df_s.drop(columns=["Close_forcast"], axis =1)
    target = df_s["Close_forcast"]
    
    df_len= df_s.shape[0]
    
    #splitting
    train_split = int (df_len *.9)
    print ("Training set size: ", train_split)
    test_split = train_split + int(df_len * .1)
    print ("test set size: ", int(df_len * .1))
    
    x_train,  x_test = features[:train_split], features[train_split:]
    y_train, y_test = target[:train_split], target[train_split:]
    
    print (x_train.shape,  x_test.shape)
    print (y_train.shape, y_test.shape)
    
    return x_train, x_test, y_train, y_test
```

```python
# cell 7
X_train, X_test, Y_train, Y_test = create_train_test_dataset(df1_processed)
```

```python
# cell 8
#training of model using Linear regression
# slope and intercept are calculated for each column
lr = LinearRegression()
lr.fit(X_train, Y_train)
print("slope values obtained are:","\n")
print (lr.coef_)
print("\n")
print("intercept value obtained is:")
print (lr.intercept_)
Y_train_pred_lr = lr.predict (X_train)

Y_test_pred_lr = lr.predict (X_test)
```

```python
# cell 9
# training carried out using svm regression algorithm
from sklearn import svm

svmr=svm.SVR()
svmr.fit(X_train, Y_train)

#model prediciton using svm

Y_train_pred_svmr = svmr.predict (X_train)

Y_test_pred_svmr = svmr.predict (X_test)


```

```python
# cell 10
#evaluation on training and testing data
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error,max_error

def metrics_result(Y_train,Y_train_pred,Y_test,Y_test_pred):

    print ("Train R_squared: ",  r2_score(Y_train, Y_train_pred),"\n")
    print ("Test R_squared: ",  r2_score(Y_test, Y_test_pred),"\n")

    print ("Train mean_squared_error: ",  mean_squared_error(Y_train, Y_train_pred),"\n")
    print ("Test mean_squared_error: ",  mean_squared_error(Y_test, Y_test_pred),"\n")

    print ("Train mean_abs_errord: ",  mean_absolute_error(Y_train, Y_train_pred),"\n")
    print ("Test mean_abs_error: ",  mean_absolute_error(Y_test, Y_test_pred),"\n")
    
    
    print ("Train_max_error: ",  max_error(Y_train, Y_train_pred),"\n")
    print ("Test_max_error: ",  max_error(Y_test, Y_test_pred),"\n")
    return None

```

```python
# cell 11
# displaying the results of both algorithms on testing and training data
# using the above function we can generate results for any regression algorithm over the time
print("results of linear regression are:")
metrics_result(Y_train,Y_train_pred_lr,Y_test,Y_test_pred_lr)
print("\n")
print("results of support vector machine regression algorithm are:")
metrics_result(Y_train,Y_train_pred_svmr,Y_test,Y_test_pred_svmr)
```

```python
# cell 12
# computing the cross validation scores using k cross validation approach
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold

kfold = KFold(n_splits=5, shuffle=True, random_state=42)

scores = cross_val_score(lr, X_train, Y_train, cv=kfold, scoring='neg_mean_squared_error')
scores1= cross_val_score(lr, X_test, Y_test, cv=kfold, scoring='neg_mean_squared_error')
scores2= cross_val_score(svmr, X_train, Y_train, cv=kfold, scoring='neg_mean_squared_error')
scores3= cross_val_score(svmr, X_test, Y_test, cv=kfold, scoring='neg_mean_squared_error')
scores=pd.DataFrame(scores)
scores1=pd.DataFrame(scores1)
scores2=pd.DataFrame(scores2)
scores3=pd.DataFrame(scores3)
print("Average Positive MSE for training data using linear regression:",-scores.mean(),"\n")

print("Average Positive MSE for testing data using linear regression:",-scores1.mean(),"\n")

print("Average Positive MSE for training data using svm regression:",-scores2.mean(),"\n")

print("Average Positive MSE for testing data using svm regression:",-scores3.mean(),"\n")

```

```python
# cell 17
# python user-defined function to plot the residuals against the predicted values
def f(a,Y_pred):
    residuals = a - Y_pred

    # Plot residuals
    plt.scatter(Y_pred, residuals)
    plt.xlabel('Predicted Values')
    plt.ylabel('Residuals')
    plt.axhline(y=0, color='r', linestyle='--')
    plt.show()
    return 0
```

```python
# cell 18
# plotting residuals vs predicted data for training data using svm regression
f(Y_train,Y_train_pred_svmr)
```

```python
# cell 19
# plotting residuals vs predicted data for testing data using svm regression
f(Y_test,Y_test_pred_svmr)
```
